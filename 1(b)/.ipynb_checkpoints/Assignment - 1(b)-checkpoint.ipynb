```python
import pandas as pd
df=pd.read_csv('iris.csv')
df.head()
#there are five features: first four are numerical features and last one nominal
```

```python
def answer_one():
    x=df.loc[:,'SepalLength':'PetalWidth']
    stdev=x.std()
    mean_df=x.mean()
    count_df=x.count()
    range_df=x.max()-x.min()
    print("STANDARD DEVIATION:\n",stdev)
    print("MEAN:\n",mean_df)
    print("COUNT:\n",count_df)
    print("MIN:\n",x.min())
    print("MAX:\n",x.max())
    print("RANGE:\n",range_df)
    print("VARIANCE:\n",stdev**2)
    print("QUARTILES:\n",x.quantile([0.25,0.5,0.75]))
answer_one()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df.hist()
```

```python
boxplot = df.boxplot(column=['SepalLength', 'SepalWidth', 'PetalLength','PetalWidth'])
```
